```python
df=pd.read_csv('ukbb_asthma_cohorts_gwas_annotation_new.csv').fillna('NA')
```

```python
df.head()
conditions = [
    (df['Asthma_CohortA'] == 1) & (df['Asthma_CohortB']==0),
    (df['Asthma_CohortA'] == 0) & (df['Asthma_CohortB']==1)]

choices = ['NA','1']
np.select(conditions, choices)
df['Asthma_CohortB']=np.select(conditions, choices)
```

```python
df.head()
conditions = [
    (df['Asthma_CohortA'] == 0) & (df['Asthma_CohortB'].str.contains('1')),
    (df['Asthma_CohortA'] == 1) & (df['Asthma_CohortB'].str.contains('NA'))
    ]
    
choices = ['NA','1']
np.select(conditions, choices)
df['Asthma_CohortA']=np.select(conditions, choices)
```

```python
df.to_csv('ukbb_asthma_cohorts_gwas_annotation_new_fixed.csv',index=False)
```
